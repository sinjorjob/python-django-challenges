### 以下の項目が繰り返されてる。

仕事内容、対象となる方、勤務地、給与、初年度年収  
 
div class="**cassetteRecruit__main**"  
   table class="**tableCondition**"
        
th class="**tableCondition__head**" ⇒項目
td class="**tableCondition__body**"　⇒内容

1番目：仕事内容    
2番目：勤務地  
3番目：給与
        
CSSセレクタ:   .cassetteRecruit__main .tableCondition

```python
#企業内容を取得
#会社名は　i.text.split(" | " )[0]　　で取得
company_list =[]
company_data =  driver.find_elements_by_css_selector(".cassetteRecruit__name")

for company in company_data:
    company_list.append(company.text.split("|")[0])
print(len(company_list))
```

```python
#勤務地と給与、初年度年収を取得
table_list =  driver.find_elements_by_css_selector(".tableCondition")
#print(len(table_list))

work_location_list = []
salary_list = []
initial_annual_income =[]


for table in table_list:
    work_location_list.append(table.find_elements_by_tag_name("td")[2].text)  #勤務地をリストに追加
    salary_list.append(table.find_elements_by_tag_name("td")[3].text)  #給与をリストに追加
    try:
        initial_annual_income.append(table.find_elements_by_tag_name("td")[4].text)    #初年度年収をリストに追加
        
    except:
        initial_annual_income.append("初年度年収のデータなし")   #初年度年収データがない場合その旨をリストに追加
    

for company, location, salary, income in zip(company_list, work_location_list, salary_list, initial_annual_income):
    print(f"会社名:{company}\n勤務地：{location}\n給与：{salary}\n初年度年収：{income}\n")
    print("="*110)
```

```python
import os
from selenium.webdriver import Chrome, ChromeOptions
import time
import pandas as pd
import sys
# Chromeを起動する関数


def set_driver(driver_path, headless_flg):
    # Chromeドライバーの読み込み
    options = ChromeOptions()

    # ヘッドレスモード（画面非表示モード）をの設定
    if headless_flg == True:
        options.add_argument('--headless')

    # 起動オプションの設定
    options.add_argument(
        '--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36')
    # options.add_argument('log-level=3')
    options.add_argument('--ignore-certificate-errors')
    options.add_argument('--ignore-ssl-errors')
    options.add_argument('--incognito')          # シークレットモードの設定を付与

    # ChromeのWebDriverオブジェクトを作成する。
    return Chrome(executable_path=os.getcwd() + "/" + driver_path, options=options)

# main処理
driver = set_driver("chromedriver.exe", False)
driver.get("https://tenshoku.mynavi.jp/")
time.sleep(5)
 
driver.execute_script('document.querySelector(".karte-close").click()')
time.sleep(5)
    
# 検索窓に入力
search_keyword = "テレワーク"
driver.find_element_by_class_name("topSearch__text").send_keys(search_keyword)
# 検索ボタンクリック
driver.find_element_by_class_name("topSearch__button").click()
```

```python
import datetime
import logging

#現在時刻の取得
now = datetime.datetime.now()
logfile = '実行ログ_' + search_keyword + "_" + now.strftime('%Y%m%d_%H%M%S') + '.log'
logging.basicConfig(filename=logfile,level=logging.DEBUG,format='%(asctime)s %(message)s', datefmt='%m/%d/%Y%H%M%S %I:%M:%S %p')



try:
    company_data =  driver.find_element_by_css_selector(".cassetteRecruit__name")

    logging.info('[正常]"test"')
    logging.info(f"キーワード：{search_keyword}")
except Exception as e:
    logging.info('[異常]異常終了しました。')
    logging.info('[異常]エラー内容:%s', str(e.args))
    logging.exception('エラー詳細: %s', e)
```

```python
count = 1
company_list =[]
work_location_list = []
salary_list = []
initial_annual_income =[]

while True:
    print(f"{count}ページ目です！")
    try:
        next_page = driver.find_element_by_class_name("iconFont--arrowLeft")
        print("次のページが見つかりました")
        #企業内容を取得
        #会社名は　i.text.split(" | " )[0]　　で取得
        
        company_data =  driver.find_elements_by_css_selector(".cassetteRecruit__name")

        company_list = [ company_list.append(company.text.split("|")[0] for company in company_data]

        print(len(company_list))

        #勤務地と給与、初年度年収を取得
        table_list =  driver.find_elements_by_css_selector(".tableCondition")
        print(f"talbe_list:{len(table_list)}")

        for table in table_list:
            work_location_list.append(table.find_elements_by_tag_name("td")[2].text)  #勤務地をリストに追加
            salary_list.append(table.find_elements_by_tag_name("td")[3].text)  #給与をリストに追加
            try:
                initial_annual_income.append(table.find_elements_by_tag_name("td")[4].text)    #初年度年収をリストに追加
        
            except:
                initial_annual_income.append("初年度年収のデータなし")   #初年度年収データがない場合その旨をリストに追加
    
        print(next_page.get_attribute("href"))
        next_page_link = next_page.get_attribute("href")
        count += 1
        driver.get(next_page_link)
        time.sleep(3)
    except:
        print("次のページが見つかりませんでした。")
        break
        
print(f"company_list:{len(company_list)}")
print(f" work_location_list:{len( work_location_list)}")
print(f"salary_list:{len(salary_list)}")
print(f"initial_annual_income:{len(initial_annual_income)}")


driver.quit()
```
